```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

**Scrapping tweets using snscrapper**


```python
tweets_list = []

for i,tweet in enumerate(sntwitter.TwitterSearchScraper('forest fire since:2021-03-01 until:2021-09-30').get_items()):
  if 'forest fire' in**Scrapping tweets using snscrapper** tweet.content:
    tweets_list.append([tweet.id, tweet.username, tweet.content, tweet.date])

tweets_df = pd.DataFrame(tweets_list, columns=['Tweet Id','Username', 'Text','Datetime' ])
tweets_df.to_csv("TwitterSats.csv")
tweets_df
```

**Getting Unique Hashtag**

```python
hashtags_set = set()
def hashtags(text):
  for word in text.split():
    if(word[0] == "#"):
      hashtags_set.add(word[1:])
tweets_df.Text.apply(hashtags)
hashtags_list = list(hashtags_set)
hashtags_list
```

**Average Tweet Length**

```python
tweet_count_per_row = tweets_df.Text.str.split().apply(len)
mean = tweet_count_per_row.mean()
mean
```
